# Google Crawling

```python
from selenium import webdriver
from urllib.request import urlopen
from selenium.webdriver.common.keys import Keys
import time
import urllib.request
import os
```

```python
def get_pic(driver, path, keyword, limit, option=1) :
    if not os.path.isdir(f"{keyword}/"):
        os.makedirs(f"{keyword}/")

    driver = webdriver.Chrome(driver)
    driver.get(path)

    search = keyword
    elem = driver.find_element_by_name("q")
    elem.send_keys(search)
    elem.send_keys(Keys.RETURN)
    count = 0

    SCROLL_PAUSE_TIME = 1
    
    new_keyword = keyword.replace(" ", '_')

    last_height = driver.execute_script("return document.body.scrollHeight")
    BT_see_more = '//*[@id="islmp"]/div/div/div/div/div[2]/span' # 
    BT_see_more2 = '//*[@id="islmp"]/div/div/div/div/div/div[2]/div[2]/input' # 결과 더 보기
    page = 0

    if option == 1:
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(SCROLL_PAUSE_TIME)
            new_height = driver.execute_script("return document.body.scrollHeight")
            time.sleep(1)
            print("page down")

            if new_height == last_height and page == 0 : #결과 더 보기
                element = driver.find_element_by_xpath(BT_see_more2)   
                element.click()
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(SCROLL_PAUSE_TIME)
                new_height = driver.execute_script("return document.body.scrollHeight")
                page += 1
                print(f"page : {page}")               

            if new_height == last_height and page == 1 : #페이지 다 내림 
                print(f"page : {page}")            
                try:
                    time.sleep(1)
                    # print("checkk point 0")                
                    # driver.find_element_by_css_selector(".mye4qd").click()
                    print("checkk point 1")
                    images = driver.find_elements_by_css_selector(".rg_i.Q4LuWd")
                    print("checkk point 2")


                    # //*[@id="islrg"]/div[1]/div[93]/a[1]/div[1]/img
                    for image in images:
                        print("checkk point 3")                     
                        try:
                            print("checkk point 4")                                             
                            image.click()
                            time.sleep(2)
                            count += 1
                            imgUrl = driver.find_element_by_xpath(f"//*[@id='islrg']/div[1]/div[{count}]/a[1]/div[1]/img").get_attribute('src')
                            lastImg = driver.find_element_by_xpath(f"//*[@id='islrg']/div[1]/div[{len(images)-1}]/a[1]/div[1]/img").get_attribute('src')
                            urllib.request.urlretrieve(imgUrl, f"{keyword}/" + new_keyword + "_" + str(count) + ".jpg")
                            print(f"Image saved: {keyword}{count}.jpg")
                            if count == limit : 
                                print("Crawling succeeded, driver shutdown.")
                                driver.close()
                            if imgUrl == lastImg :
                                print("Crawling succeeded, driver shutdown.")
                                driver.close()

                        except:
                            pass
                except:
                    print("ERROR BREAK!!")
                    break
            last_height = new_height
    elif option == 2 :
        while True:
            driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
            time.sleep(SCROLL_PAUSE_TIME)
            new_height = driver.execute_script("return document.body.scrollHeight")
            time.sleep(1)
            print("page down")

            if new_height == last_height and page == 0: # 나머지 검색결과는 내가 찾고 있는 항목이 아닐 수도 있습니다.
                print("if 1 start")
                element = driver.find_element_by_xpath(BT_see_more)   
                element.click()
                page += 1
                time.sleep(1)
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(SCROLL_PAUSE_TIME)  
                new_height = driver.execute_script("return document.body.scrollHeight")            
                print(f"page : {page}")

            if new_height == last_height and page == 1 : #결과 더 보기
                element = driver.find_element_by_xpath(BT_see_more2)   
                element.click()
                driver.execute_script("window.scrollTo(0, document.body.scrollHeight);")
                time.sleep(SCROLL_PAUSE_TIME)
                new_height = driver.execute_script("return document.body.scrollHeight")
                page += 1
                print(f"page : {page}")               

            if new_height == last_height and page == 2 : #페이지 다 내림 
                print(f"page : {page}")            
                try:
                    time.sleep(1)
                    # print("checkk point 0")                
                    # driver.find_element_by_css_selector(".mye4qd").click()
                    print("checkk point 1")
                    images = driver.find_elements_by_css_selector(".rg_i.Q4LuWd")
                    print("checkk point 2")                

                    # //*[@id="islrg"]/div[1]/div[93]/a[1]/div[1]/img
                    for image in images:
                        print("checkk point 3")                     
                        try:
                            print("checkk point 4")                                             
                            image.click()
                            time.sleep(2)
                            count += 1
                            print(count)
                            imgUrl = driver.find_element_by_xpath(f"//*[@id='islrg']/div[1]/div[{count}]/a[1]/div[1]/img").get_attribute('src')
        #                     print(imgUrl)
                            urllib.request.urlretrieve(imgUrl, f"{keyword}/" + new_keyword + "_" + str(count) + ".jpg")
                            print(f"Image saved: {keyword}{count}.jpg")
                            if count == limit : 
                                print("Crawling succeeded, driver shutdown.")
                                driver.close()            
                            if imgUrl == lastImg :
                                print("Crawling succeeded, driver shutdown.")
                                driver.close()                            
                        except:
                            pass
                except:
                    print("ERROR BREAK!!")
                    break
            last_height = new_height
```

#### option = 1,  "결과 더 보기"로 바로 이어지는 코드
#### option = 2,  "나머지 검색결과는 내가 찾고 있는 항목이 아닐 수도 있습니다. 검색결과 더보기" 로 이어질때 

```python
driver = "/Users/werther/chromedriver" #chrome driver 위치
path = 'https://www.google.co.kr/imghp?hl=ko&ogbl'
keyword = '강호동' 
limit = 400 # 추출할 데이터의 한도치
option = 1

get_pic(driver, path, keyword, limit, option)        
```
